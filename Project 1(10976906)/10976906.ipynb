### *Filling null values with Median method*


```python
med= df_raw.median()
df_final = df_raw.fillna(med)
df_final
```

### *Using the median to fill null values in Panas is a common strategy when dealing with numerical data. The median is a measure of central tendency that represents the middle value of a dataset. Filling null values with the median can help to preserve the shape of the distribution of the data Calculating the median is relatively quick and does not require as much computational resources as other methods, such as interpolation or regression.*

```python
df_final.info()
```

```python
df_final.isna().sum()
```

### *Mode, Mean and Median*


```python
mean_df_final = df_final.iloc[:, 1:].mean() 
mode_df_final = df_final.iloc[:, 1:].mode().iloc[0]
median_df_final = df_final.iloc[:, 1:].median()



print("Mean:")
print(mean_df_final)
print("\nMode:")
print(mode_df_final)

```

```python
print("\nMedian:")
print(median_df_final)
```

### Data Visualization for 'Fixed Acidity'

```python
mean1 = df_final['fixed acidity'].mean()

# Count values above and below mean
above_mean_count = df_final[df_final['fixed acidity'] > mean1]['fixed acidity'].count()
below_mean_count = df_final[df_final['fixed acidity'] < mean1]['fixed acidity'].count()

# Create pie chart
labels = ['Above Mean', 'Below Mean']
sizes = [above_mean_count, below_mean_count]
colors = ['red', 'yellow']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Fixed Acidity Above and Below Mean')
plt.show()
```

### *Heatmap* 
---

#### *Heatmaps are commonly used to visualize correlations between each columns. Correlation coefficients can be visualized as a color gradients.This allows us to easily identify variables that are strongly correlated or anti-correlated.*

```python
col = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']

# Compute correlation matrix
corr_data = df_final[col].corr()

fig,ax = plt.subplots(figsize=(12, 8)) ## Set figure size

# Plot correlation heatmap
data_heatmap = sns.heatmap(corr_data, annot=True, ax=ax,  cmap='inferno')

data_heatmap.set_xticklabels(data_heatmap.get_xticklabels(), rotation=30)
```

### *Hidden Inferences*
---

#### From the Heatmap above:

#### Density is well correlated with fixed acidity
#### Fixed acidity is anti-correlated with pH
#### Density is anti-correlated with Alcohol
